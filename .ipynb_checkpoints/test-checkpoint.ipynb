Simple bisection for testing

```python
def bisection(f, a, b, eps):
    print("a = ", a, " b = ", b)
    step = 1
    condition = True
    while condition:
        mid = (a + b) / 2
        #print('Iteration %d: mid = %f, f(mid) = %f' % (step, mid, f(mid)))

        if f(a) * f(mid) < 0:
            b = mid
        else:
            a = mid
        
        step = step + 1
        condition = abs(f(mid)) > eps
    
    print('Root is: %f' % mid)
```

```python
import numpy as np
def f(x):
    if x < -3:
        return -x
    if x < -1 and x>-3:
        return -2*x
    else:
        return -(x**2 + 4*x + 3)
```

```python
x=np.linspace(-10, 5, 100)
fun = []
for i in x:
    fun.append(f(i))
plt.plot(x, fun)
plt.axhline(y=0, color='r')
```

```python
for i in [-5, -4, -3, -2.5, -2, -1.5]:
    for j in [-0.5, 0, 0.5, 1, 1.5, 2]:
        bisection(f, i, j, 1e-9)
```
